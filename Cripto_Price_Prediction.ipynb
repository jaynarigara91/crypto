```python
import numpy as np
import pandas as pd
import matplotlib.dates as mdates
import yfinance as yf
import datetime as dt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.layers import Dense, LSTM, Dropout
from tensorflow.keras.models import Sequential
from sklearn.metrics import mean_absolute_error,mean_squared_error, r2_score
import matplotlib.pyplot as plt

# List of cryptocurrencies
crypto_currencies = ['BTC-USD', 'LTC-USD', 'ETH-USD', 'XMR-USD']

# Define the prediction days and training data range
start = dt.datetime(2020, 1, 1)
end = dt.datetime.now()
prediction_day = 60

for crypto_currency in crypto_currencies:
    # Fetch data using yfinance
    data = yf.download(crypto_currency, start=start, end=end)

    # Scale the data
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(data['Close'].values.reshape(-1, 1))

    # Prepare the training dataset
    X_train, Y_train = [], []
    for x in range(prediction_day, len(scaled_data)):
        X_train.append(scaled_data[x - prediction_day:x, 0])
        Y_train.append(scaled_data[x, 0])

    X_train, Y_train = np.array(X_train), np.array(Y_train)
    X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))

    # Build the LSTM model
    model = Sequential()
    model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
    model.add(Dropout(0.2))
    model.add(LSTM(units=50, return_sequences=True))
    model.add(Dropout(0.2))
    model.add(LSTM(units=50))
    model.add(Dropout(0.2))
    model.add(Dense(units=1))

    model.compile(optimizer='adam', loss='mean_squared_error')
    model.fit(X_train, Y_train, epochs=25, batch_size=32)

    # Test dataset
    test_start = dt.datetime(2023, 1, 1)
    test_end = dt.datetime.now()

    # Fetch test data using yfinance
    test_data = yf.download(crypto_currency, start=test_start, end=test_end)
    actual_prices = test_data['Close'].values

    # Prepare the model inputs for prediction
    total_dataset = pd.concat((data['Close'], test_data['Close']), axis=0)
    model_inputs = total_dataset[len(total_dataset) - len(test_data) - prediction_day:].values
    model_inputs = model_inputs.reshape(-1, 1)
    model_inputs = scaler.fit_transform(model_inputs)

    # Prepare the test dataset
    X_test = []
    for x in range(prediction_day, len(model_inputs)):
        X_test.append(model_inputs[x - prediction_day:x, 0])

    X_test = np.array(X_test)
    X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

    # Predict the prices
    prediction_prices = model.predict(X_test)
    prediction_prices = scaler.inverse_transform(prediction_prices)

    # Generate dates for all data and test data
    all_dates = data.index  
    test_dates = test_data.index  

    plt.figure(figsize=(12, 6))
    plt.plot(all_dates, data['Close'].values, color='Blue', label='Actual Prices')
    # plt.plot(test_dates, actual_prices, color='red', label='Actual Prices')
    plt.plot(test_dates, prediction_prices, color='red', label='Predicted Prices')
    plt.title(f'{crypto_currency} Price Prediction') 
    plt.xlabel('Year')  
    plt.ylabel('Price')  
    plt.legend(loc='upper left')
    plt.gca().xaxis.set_major_locator(mdates.YearLocator())  
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))  
    plt.grid(True)  
    plt.tight_layout()  
    plt.show()


    # Predict the next day's price
    real_data = [model_inputs[len(model_inputs) + 1 - prediction_day:len(model_inputs) + 1, 0]]
    real_data = np.array(real_data)
    real_data = np.reshape(real_data, (real_data.shape[0], real_data.shape[1], 1))

    prediction = model.predict(real_data)
    prediction = scaler.inverse_transform(prediction)
    print(f"Next predicted price for {crypto_currency}: {prediction}")
    
    mse = mean_squared_error(actual_prices, prediction_prices)
    print(f'Root Mean Squared Error (RMSE): {mse}')
    
    Accuracy = r2_score(actual_prices, prediction_prices)
    print(f"Accuracy : {Accuracy}")
```

```python
import numpy as np
import pandas as pd
import yfinance as yf
import datetime as dt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import xgboost as xgb
import plotly.graph_objects as go

# Load the cryptocurrency data
crypto_currency = 'BTC-USD'
start = dt.datetime(2020, 1, 1)
end = dt.datetime.now()
data = yf.download(crypto_currency, start=start, end=end)

# Scale the data
scaler = MinMaxScaler(feature_range=(0, 1))
data['Close'] = scaler.fit_transform(data['Close'].values.reshape(-1, 1))

# Create lag features for time series forecasting
prediction_day = 60  # Number of days to look back
X = []
Y = []

for i in range(prediction_day, len(data)):
    X.append(data['Close'].values[i - prediction_day:i])
    Y.append(data['Close'].values[i])

X = np.array(X)
Y = np.array(Y)

# Flatten the features to use with XGBoost
X = X.reshape(X.shape[0], X.shape[1])

# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, shuffle=False)

# Convert to DMatrix for better performance
dtrain = xgb.DMatrix(X_train, label=Y_train)
dtest = xgb.DMatrix(X_test, label=Y_test)

# Set parameters
params = {
    'objective': 'reg:squarederror',
    'eval_metric': 'rmse',
    'max_depth': 6,
    'learning_rate': 0.01,
    'subsample': 0.8,
    'colsample_bytree': 0.8
}

# Train an XGBoost model with early stopping
eval_set = [(dtrain, 'train'), (dtest, 'test')]
model = xgb.train(
    params=params,
    dtrain=dtrain,
    num_boost_round=1000,
    evals=eval_set,
    early_stopping_rounds=50,
    verbose_eval=True
)

# Make predictions
predictions = model.predict(dtest)

# Inverse scale the predictions and the actual values
predictions = scaler.inverse_transform(predictions.reshape(-1, 1))
actual_prices = scaler.inverse_transform(Y_test.reshape(-1, 1))

# Function to predict future prices for the given interval
def predict_future_prices(model, interval):
    last_sequence = data['Close'].values[-prediction_day:]
    future_predictions = []
    for _ in range(interval):
        dmatrix = xgb.DMatrix(last_sequence.reshape(1, -1))
        pred = model.predict(dmatrix)
        future_predictions.append(pred[0])
        last_sequence = np.append(last_sequence[1:], pred[0])
    future_predictions = scaler.inverse_transform(np.array(future_predictions).reshape(-1, 1))
    return pd.date_range(start=data.index[-1], periods=interval + 1, freq='D')[1:], future_predictions.flatten()

# Plot the actual vs predicted prices with interactive buttons
fig = go.Figure()

# Add the initial actual prices and predictions for 1-day forecast
fig.add_trace(go.Scatter(x=data.index[-len(actual_prices):], y=actual_prices.flatten(), mode='lines', name='Actual Price'))
future_dates, future_prices = predict_future_prices(model, 1)
fig.add_trace(go.Scatter(x=future_dates, y=future_prices, mode='lines', name='Predicted Price (1 Day)'))

# Buttons for different intervals
buttons = [
    dict(
        label="1 Day",
        method="update",
        args=[{"x": [data.index[-len(actual_prices):], future_dates], 
               "y": [actual_prices.flatten(), future_prices]},
              {"title": f"{crypto_currency} Price Prediction: 1 Day"}],
    ),
    dict(
        label="1 Week",
        method="update",
        args=[{"x": [data.index[-len(actual_prices):], *predict_future_prices(model, 7)[0:1]], 
               "y": [actual_prices.flatten(), *predict_future_prices(model, 7)[1:2]]},
              {"title": f"{crypto_currency} Price Prediction: 1 Week"}],
    ),
    dict(
        label="1 Month",
        method="update",
        args=[{"x": [data.index[-len(actual_prices):], *predict_future_prices(model, 30)[0:1]], 
               "y": [actual_prices.flatten(), *predict_future_prices(model, 30)[1:2]]},
              {"title": f"{crypto_currency} Price Prediction: 1 Month"}],
    ),
    dict(
        label="3 Months",
        method="update",
        args=[{"x": [data.index[-len(actual_prices):], *predict_future_prices(model, 90)[0:1]], 
               "y": [actual_prices.flatten(), *predict_future_prices(model, 90)[1:2]]},
              {"title": f"{crypto_currency} Price Prediction: 3 Months"}],
    ),
    dict(
        label="1 Year",
        method="update",
        args=[{"x": [data.index[-len(actual_prices):], *predict_future_prices(model, 365)[0:1]], 
               "y": [actual_prices.flatten(), *predict_future_prices(model, 365)[1:2]]},
              {"title": f"{crypto_currency} Price Prediction: 1 Year"}],
    )
]

# Add buttons to the layout
fig.update_layout(
    updatemenus=[dict(type="buttons", direction="left", buttons=buttons, x=0.5, y=1.2, xanchor='center', yanchor='top')],
    title=f"{crypto_currency} Price Prediction",
    xaxis_title="Date",
    yaxis_title="Price",
    template="plotly_white"
)

fig.show()
```
